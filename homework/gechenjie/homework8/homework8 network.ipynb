### 绘制www网络的出度分布、入度分布

```python
in_dc = nx.in_degree_centrality(g)
out_dc= nx.out_degree_centrality(g)
```

```python
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot(1, 3, 1)
plt.hist(in_dc.values(), bins = 3)
plt.xlabel('$in_degree_Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)

ax = plt.subplot(1, 3, 2)
plt.hist(out_dc.values(), bins = 3)
plt.xlabel('$out_degree_Centrality$', fontsize = 20)

plt.tight_layout()
plt.show()
```

### 使用BA模型生成节点数为N、幂指数为γ的网络
### 计算平均路径长度d与节点数量的关系

```python
import networkx as nx
import matplotlib.pyplot as plt
Ns = [i*10 for i in [1, 10, 100, 1000]]
ds = []
for N in Ns:
    BA= nx.random_graphs.barabasi_albert_graph(N,2)
    d = nx.average_shortest_path_length(BA)
    ds.append(d)
    
pos = nx.spring_layout(BA)          #定义一个布局，此处采用了spring布局方式
nx.draw(BA,pos,with_labels=False,node_size = 30)  #绘制图形
plt.show()
```

```python
plt.plot(Ns, ds, 'r-o')
plt.xlabel('$N$', fontsize = 20)
plt.ylabel('$<d>$', fontsize = 20)
plt.xscale('log')
plt.show()
```
