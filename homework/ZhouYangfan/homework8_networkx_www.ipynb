#### 计算网络中的节点数量和链接数量

```python
%matplotlib inline
import networkx as nx
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
n = 0
with open ('/Users/apricot/documents/github/bigdata/www.dat.gz.txt') as f:
    for line in f:
        x, y = line.rstrip().split(' ')
        G.add_edge(x,y)
```

### 作业2：
#### 绘制www网络的出度分布、入度分布

```python
in_degree=nx.in_degree_centrality(G)
out_degree=nx.out_degree_centrality(G)
```

```python
fig = plt.figure(figsize=(15, 4),facecolor='white')
ax = plt.subplot(1, 3, 1)
plt.hist(in_degree.values(), bins = 3)
plt.xlabel('$In_degree \, Centrality$', fontsize = 20)
plt.ylabel('$Frequency, \, F$', fontsize = 20)

ax = plt.subplot(1, 3, 2)
plt.hist(out_degree.values(), bins = 3)
plt.xlabel('$Out_degree\, Centrality$', fontsize = 20)
plt.tight_layout() 
plt.show()
```

#### 使用BA模型生成节点数为N、幂指数为γ的网络，绘制平均路径长度d与节点数量的关系

```python
import networkx as nx
import matplotlib.pyplot as plt

Ns = [i*10 for i in [1, 10, 100, 1000]]
ds = []
for N in Ns:
    print N
    BA= nx.random_graphs.barabasi_albert_graph(N,2)
    d = nx.average_shortest_path_length(BA)
    ds.append(d)

pos = nx.spring_layout(BA) 
nx.draw(BA,pos,with_labels=False,node_size = 30)  #绘制图形
plt.show()
```

```python
plt.plot(Ns, ds, 'r-o')
plt.xlabel('$N$', fontsize = 20)
plt.ylabel('$<d>$', fontsize = 20)
plt.xscale('log')
plt.show()
```
