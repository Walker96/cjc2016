# while循环

```python
j = -2
while j <10:
    print j
    j+=2 # avoid dead loop
```

```python
j = 1
while j <10:
    if j%3 != 0: 
        print j**2
    j+=2 # avoid dead loop
```

```python
j = 1
while j <10:
    if j == 5:
        break
    if j%2 != 0: 
        print j**2
    j+=2 # avoid dead loop
```
